# Based on question - Colouring based on values

```python
import numpy as np 

month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
         'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [1100, 1200, 1300, 1350, 1400, 1387,
         1499, 1574, 1611, 1644, 1587, 1721]

sales2 = np.array(sales)
month2 = np.array(month)

mask1 = sales2>1300
mask2 = sales2<=1300

plt.bar(month2[mask2], sales2[mask2], color = 'red')
plt.bar(month2[mask1], sales2[mask1], color = 'green')


plt.show()
```

```python
month2 = np.array(month)
sales2 = np.array(sales)


cc =['color']*len(sales2)   
for n,val in enumerate(sales2):
    if val>1300:
        cc[n]='green'
    else:
        cc[n]='red'    

plt.bar(month2, sales2, color = cc)
plt.show()
```

## Scatter Plot

```python
summer2016 = pd.read_csv('summer2016.csv')
summer2016.head()
```

Continuous variables

## Data Types

### Matplotlib

```python
plt.scatter(summer2016['Height'],summer2016['Weight'], s=2); #change size of bar
```

### Let's add labels

```python
plt.scatter(summer2016['Height'],summer2016['Weight'], s=2)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Olympic Athletes")
#change size of bar
```

# Adding Vertical Lines and limiting the axes

```python
plt.subplots(figsize=(10,5)) #change the size of the figure - > (width, height)
plt.scatter(summer2016["Height"], summer2016["Weight"], s=2)


plt.xlim(left=0,right=1000) #To specify range of axis to be shown (optional parameters)
#You can choose to ignore any of the parameters above (e.g left or right) and it'll automatically use the default limit


plt.axvline(x=100, color="green", label="vert") #to include a vertical line (e.g targets)
plt.axhline(y=100, color="red", label="horz")  #to include a horizontal line (e.g. targets)


plt.xticks(rotation=45) #rotate the values shown on the x-axis
plt.yticks(rotation=45) #rotate the values shown on the y-axis

plt.xlabel("Height (cm)") #label x-axis
plt.ylabel("Weight (kg)") #label y-axis
plt.title("Olympic Athletes")

plt.legend() #show labels 

plt.savefig("sample_image.png")  #normal quality
plt.savefig("sample_image2.png", dpi=300)  #high quality


plt.show()
```

### Seaborn

```python
sns.scatterplot(data=summer2016, x='Height',y='Weight', hue='Sex');

#add another column
```

```python
sns.scatterplot(data=summer2016, x='Height',y='Weight', hue='Sex', size='Medal');  

#add another column
```

### Data Dictionary

**survived** -->  0 = No, 1 = Yes <br>
**pclass** --> Ticket class	--> 1 = 1st, 2 = 2nd, 3 = 3rd <br>
**sex** --> gender <br>
**Age**	--> 	Age in years <br>
**sibsp** --> # of siblings / spouses aboard the Titanic <br>
**parch** --> # of parents / children aboard the Titanic <br>
**fare** --> Passenger fare <br>
**embarked** --> Port of Embarkation C = Cherbourg, Q = Queenstown, S = Southampton <br>
**class** --> similar to pclass <br>
**who** --> gender <br>
**adult_male** --> adult_male <br>
**deck** --> deck   <br>
**embark_town** --> Port of Embarkation <br>
**alive** --> alive <br>
**alone** --> Travelling alone <br>
